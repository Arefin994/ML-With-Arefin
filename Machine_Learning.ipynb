### Feature Extraction

mapping from textual data to real values vectors

BOW = Bag of words = number of unique words in the text collection

TF-IDF = Term Frequency = number of times a word appared in the collection

IDF = - Inverse Document Frequency = log(N/n) ,  where N = the number of documents, n = the number of documwent in which the term t has appared on.

*** The IDF of a rare word is high and IDF of a frequent word is low

TF-IDF value of a term = TF x IDF

Stemming = the process of reducing a word to it's root word

Ex-player, playing, played --> play

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
  if isinstance(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content
                      if not word in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```
