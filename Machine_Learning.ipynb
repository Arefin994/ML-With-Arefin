### Feature Extraction

mapping from textual data to real values vectors

BOW = Bag of words = number of unique words in the text collection

TF-IDF = Term Frequency = number of times a word appared in the collection

IDF = - Inverse Document Frequency = log(N/n) ,  where N = the number of documents, n = the number of documwent in which the term t has appared on.

*** The IDF of a rare word is high and IDF of a frequent word is low

TF-IDF value of a term = TF x IDF

Stemming = the process of reducing a word to it's root word

Ex-player, playing, played --> play

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
  if isinstance(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content
                      if not word in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
newsData['content'] = newsData['content'].apply(stemming)
```

#### Cross Validation

in SciKit Learn library, we have a cross validation function

In cross validation the entire data is used 5 times like this -

image.png

Here different parts of the data is used as the test data to see the accuracy score. After that it takes the mean values of the data set for using a model.

After checking the models we can compare and see which model is the best for our datasets.

Let's see it in code -

```python
#importing the deps
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
# let's test which model is the best for the diabetes dataset
data = pd.read_csv('/content/diabetes.csv')
```

```python
X = data.drop(columns = 'Outcome', axis=1)
Y = data['Outcome']
```

```python
# Standardized the data
scalar = StandardScaler()
```

```python
scalar.fit(X)
```

```python
std_data = scalar.transform(X)
```

REMEMBER :  We must need to preprocess the data before doing the cross validation!!!

```python
X = std_data
Y = data['Outcome']

# Logistic Regression
logreg = LogisticRegression()
logreg_cv = cross_val_score(logreg, X, Y, cv=5)
print("Logistic Regression Accuracy:", logreg_cv.mean())

# Support Vector Machine
svc = SVC()
svc_cv = cross_val_score(svc, X, Y, cv=5)
print("Support Vector Machine Accuracy:", svc_cv.mean())

# Random Forest
rf = RandomForestClassifier()
rf_cv = cross_val_score(rf, X, Y, cv=5)
print("Random Forest Accuracy:", rf_cv.mean())

# Find the best model
models = {'Logistic Regression': logreg_cv.mean(), 'Support Vector Machine': svc_cv.mean(), 'Random Forest': rf_cv.mean()}
best_model = max(models, key=models.get)
print(f"The best model is {best_model} with an accuracy of {models[best_model]}")
```

### OverFitting

when a model learns the training data too well, capturing noise and details that don’t generalize to new data.

image.png

we will know the model is overfitted if we get very high training data accuracy but low test accuracy.

image.png

If the accuracy of test and train data is almost same we can say the model is not overfitted



Reasons for over fitting -


1.   less data
2.   Increased compxity of the model
3.    More number of layers in the nural network


Preventing Overfitting -

1.  using more data
2.  reduce the number of layer in the nural network
3.  Early Stopping
4.  Bias - Varianc Tradeoff
5.   Use Dropouts

### Underfitting

when a model does not learn enough from the data

image.png

we will know the model is underfitted if the training data is very low

Reasons for underfitting

1. Choosing a wrong model
2. Less coplexity of the model
3. Less varince but high bias

Prevent underfitting

1. Choosing the correct model
2. increse the complexity of the model
3. more number of parameters to the model
4. Bias - Variance Tradeoff
